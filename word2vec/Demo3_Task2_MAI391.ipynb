```python
#1. Load tất cả các dòng vào data frame Pandas
import pandas as pd
import re
from nltk import ngrams
from gensim.models import Word2Vec

df = pd.read_csv("/content/drive/MyDrive/DATASET_TASK2_MAI391/Binh Ngo Dai Cao.txt",sep="/", names=["row"]).dropna()
df.head(7)
```

```python
#2. Xử lý từng dòng: Xóa số dòng, bỏ chấm chấm, phẩy,  … các dấu dư thừa.
import pandas as pd
import re

def transform_row(row):
    # Xóa số dòng ở đầu câu
    row = re.sub(r"^[0-9\.]+", "", row)
    
    # Xóa dấu chấm, phẩy, hỏi ở cuối câu
    row = re.sub(r"[\.,\?]+$", "", row)
    
    # Xóa tất cả dấu chấm, phẩy, chấm phẩy, chấm thang, ... trong câu
    row = row.replace(",", " ").replace(".", " ") \
        .replace(";", " ").replace("“", " ") \
        .replace(":", " ").replace("”", " ") \
        .replace('"', " ").replace("'", " ") \
        .replace("!", " ").replace("?", " ")
    
    row = row.strip()
    return row 

df["row"] = df.row.apply(transform_row)
df.head(10)
```

```python
#3.Tách từ sử dụng ngram

from nltk import ngrams

def kieu_ngram(string, n=1):
    gram_str = list(ngrams(string.split(), n))
    return [ " ".join(gram).lower() for gram in gram_str ]

df["1gram"] = df.row.apply(lambda t: kieu_ngram(t, 1))
df["2gram"] = df.row.apply(lambda t: kieu_ngram(t, 2))

df.head(10)
```

```python
#4.Combine data và word2vec với Gensim Python
#https://github.com/duyet/truyenkieu-word2vec
link bai chi tiet
```

```python
df["content"] = df["1gram"] + df["2gram"]
df.head(10)
```

```python
train_data = df.content.to_list()
len(train_data)
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = Word2Vec(train_data, size=100, min_count=0 , workers=4 , sg=1)
```
